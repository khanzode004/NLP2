```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import random
import re
```

Enter books URL seprated by commas.


1.   Go to https://www.gutenberg.org/

2.   Select any book and click on "Read this book online:HTML"

3.   Copy thew link for the given book and paste it in the input box.

Multiple URLs can be entered but make sure to **seprate them by commas**

EXAMPLE: https://www.gutenberg.org/files/67185/67185-h/67185-h.htm,  https://www.gutenberg.org/files/67164/67164-h/67164-h.htm


```python
url_inp=input()
url_inp=url_inp.split(',')
```

```python
books_array=[]
alphabet = ["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"];
for url in url_inp:
  tempStr=""
  r = requests.get(url)
  soup = BeautifulSoup(r.content, 'html5lib')
  p_array=soup.find_all('p', {'style': '', 'class':''} )
  for i in p_array:
    tempStr+=re.sub("\[.*?[0-9]*\]", " ", i.get_text().replace('\n', ' '))
  books_array.append(tempStr)
```

```python
book_series = pd.Series(books_array, index = alphabet[:len(books_array)])
books_df = pd.DataFrame(book_series)
```

```python
randomsample={}
temp_array=[]
for i in books_df.index:
  arr = books_df[0][i].split()
  tempData={}
  x=random.sample(range(1, len(arr)-100), 200)
  for count, j in enumerate(x):
    tempData[count]=' '.join([str(elem) for elem in arr[j:j+100]])
  randomsample[i]=tempData
```

```python
books_sample = pd.DataFrame.from_dict(randomsample)
```
